```python
data = pd.read_csv('../data/new_locotamerzoa_tweets.csv').text.tolist()
```

```python
def filter_rt(string):
    return string.startswith("RT")

def process_tweet_text(string):
    string = re.sub('\"+', '', string) # Remove double quotes
    string = re.sub("\'+", '', string) # Remove quotes
    string = re.sub("â€¦", '', string) # remove chars
    string = re.sub('http[s]*\:\S+', '', string) # Remove URLs
    string = re.sub(r'@[a-zA-Z0-9_]+', '', string)  # Remove @ mentions
    string = re.sub(r'([a-zA-Z0-9_]{1})([\.\!\?]{1})', r'\1 \2', string) # Add space to ..
    string = re.sub(r'[0-9]+\:[0-9]+', '', string) # Remove times
    string = re.sub(r'([\U00002702-\U000027B0,\U000024C2-\U0001F251,\U0001F600-\U0001F64F,\U0001F300-\U0001F5FF,\U0001F680-\U0001F6FF,\U0001F1E0-\U0001F1FF]+)', r' \1 ', string) # add spaces in emojis

    # Handle common HTML entities
    string = re.sub(r'&lt;', '<', string)
    string = re.sub(r'&gt;', '>', string)
    string = re.sub(r'&amp;', '&', string)

    return " ".join(string.strip().split())

def filter_data(data):
    data = [el for el in data if not filter_rt(el)]
    data = [process_tweet_text(el) for el in data]
    data = [el for el in data if len(el) > 0]
    return data
```

```python
filter_data(data)
```

```python
filtered = filter_data(data)
```

```python
with open('tweets.txt', 'w') as file:
    file.write("\n".join(filtered))
```

```python
from textgenrnn import textgenrnn

textgen = textgenrnn()
textgen.reset()
textgen.train_from_file("tweets.txt", new_model=True, word_level=True, num_epochs=13, gen_epochs=13, dropout=0.5)
```
